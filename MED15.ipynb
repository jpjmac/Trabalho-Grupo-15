## 2.Homologias por Blast

```python
from Bio.Blast import NCBIXML
from Bio.Blast import NCBIWWW
from Bio import Entrez

#Como o gene apresenta 80040 ácidos nucleicos, esta é muito grande e o Blast não está a responder. Para tal, corremos o Blast só para o primeiro cds, sendo que os restantes resultados devem ser idênticos
Entrez.email="freitasjoao14@gmail.com"
my_seq=open("sequencia.txt").read()
print(my_seq)

#guarda o resultado no ficheiro xml
result_handle=NCBIWWW.qblast("blastn","nt",my_seq,hitlist_size=5,alignments=5,megablast=True)
with open("my_blast4.xml","w") as out_handle:
    out_handle.write(result_handle.read())
    
handle1=open("my_blast4.xml")
blast_records=NCBIXML.parse(handle1)

c=0
e_value=0.04
for blast_record in blast_records:
    for alignment in blast_record.alignments:
        for hsp in alignment.hsps:
            if (c<1):
                print("****Alignment****")
                print("sequencia:",alignment.title)
                print("tamanho:",alignment.length)
                print("score:",hsp.score)
                print("gaps:",hsp.gaps)
                print("e value:",hsp.expect)
                print(hsp.query[0:100]+"...")
                print(hsp.match[0:100]+"...")
                print(hsp.sbjct[0:100]+"...")
                c=c+1
            else:
                c=0
                break
```
